$$ \sqrt{x^2 + 4} $$

```python
import sympy as sp

sp.var("x")
def f(x):
    return (x**2 +4 )**0.5
dx = sp.diff(f(x), x) 
pc = sp.solve(dx, x)
print("El punto crítico está en " + str(pc))

x = np.linspace(-10,10,1000)
y = (x**2 + 4)**0.5
plt.figure()
plt.plot(x,y)
plt.grid()
```
